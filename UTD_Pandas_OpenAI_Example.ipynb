```python
import re

def nameclassifier_v3(name):
    """
    This function classifies a name into one of three categories:
    - Person
    - Company
    - Church or University
    - Other

    Args:
      name: The name to be classified.

    Returns:
      A string representing the category of the name.
    """

    # Convert the name to lowercase and strip whitespace
    name = name.lower().strip()

    # Check for company indicators
    company_indicators = ["inc", "ltd", "llc", "corporation", "corp"]
    if any(name.endswith(indicator) for indicator in company_indicators):
        return "Company"

    # Check for common title or honorifics that might indicate a person's name
    person_indicators = ["mr.", "mrs.", "ms.", "dr.", "prof."]
    if any(name.startswith(indicator) for indicator in person_indicators):
        return "Person"

    # Check for church and university indicators
    if re.search(r"\bchurch\b|\buniversity\b", name):
        return "Church or University"

    # Check if the name follows typical person name patterns (First Last)
    # Using regular expression for basic pattern matching
    if re.match(r"^[a-z]+ [a-z]+$", name):
        return "Person"

    # Otherwise, the name is classified as "Other".
    return "Other"

```

```python
ownership["Ownername_Classify"] = ownership["OwnerName"].astype(str).apply(nameclassifier_v3)
```
